```python
from costhetaa import costhetaa
%matplotlib inline
massx=np.arange(1,4,0.501)# M200,per unit 10**10Msun
#masstest3=10**massx/1e10
masstest3=10**massx
y_z0,error_z0=costhetaa(masstest3,showG_z0,w_z0,v_z0,halo_z0)
y_z0p5,error_z0p5=costhetaa(masstest3,showG_z0p5,w_z0p5,v_z0p5,halo_z0p5)
y_z1,error_z1=costhetaa(masstest3,showG_z1,w_z1,v_z1,halo_z1)
y_z3,error_z3=costhetaa(masstest3,showG_z3,w_z3,v_z3,halo_z3)
```
